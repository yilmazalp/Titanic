### Analysis of Titanic Dataset

Aim of data analysis is viewing correlation between marriage of women, being men, being children and surviving. Also, it was seen passengers who live in same cabin which means it was analyzed people from same family. 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import Imputer
from sklearn.cross_validation import train_test_split 
```

It was merged train and test set. 

```python
dataset_train = pd.read_csv("/home/nbuser/library/train.csv")
dataset_test = pd.read_csv("/home/nbuser/library/test.csv")
dataset = pd.concat([dataset_train, dataset_test])
```

```python
dataset.head(10)
```

'Embarked', 'Fare', 'Parch', 'PassengerId', 'SibSp', 'Ticket' features are unnecessary for this analysis. So, this columns were not used. 'Name' and 'Cabin' features are important for viewing families who live in same cabin. 

```python
Name_data = dataset['Name']
Cabin_data = dataset['Cabin']
Pclass_data = dataset['Pclass']
Age_data = dataset['Age']
Sex_data = dataset['Sex']
Survived_data = dataset['Survived']
```

```python
name_list = []
passenger = []
passenger_list = []
Cabin_list = []
Pclass_list = []
Age_list = []
Survived_list = []
Sex_data_list = []
title_list = []
```

```python
for name, cabin, class_, age, sex, status in zip(Name_data, Cabin_data, Pclass_data, Age_data, Sex_data, Survived_data):
    name_list.append(name)
    Cabin_list.append(cabin)
    Pclass_list.append(class_)
    Age_list.append(age)
    Survived_list.append(status)
    Sex_data_list.append(sex)
    
```

Name list was created for all passengers in the ship.

We need titles to research correlation between marriage and survived. 'Mr', 'Mrs', and 'Miss' titles are important for this research. So, title list was created. 

```python
for index in range(len(name_list)):
    passenger.append(name_list[index].split(', '))
    passenger_list.append(passenger[index][0])
    s = passenger[index][1].split('.')
    title_list.append(s[0])
```

If number of travelling people in same cabin is more than 1, the people in cabin are from same family. So, 'families' was created. 

```python
last_dataset = pd.DataFrame({'count' : new_data.groupby( [ "Name", "Cabin"] ).size()}).reset_index()
family_count_list = last_dataset.iloc[:, -1]
family_name_list = last_dataset.iloc[:, 0]
families = []

last_dataset
family_name_list
for count in range(len(family_count_list)):
    if family_count_list[count] > 1:
        families.append(family_name_list[count])
        
```

'last_dataset' was created to see that how many people from a family in same cabin. 

```python
last_dataset
```

Families in the ship and members of them. 

```python
family_data = []
for f_index in range(len(families)):
    family_data.append(new_data.loc[new_data['Name'] == families[f_index]])
    
family_data
```

In these familes, it was researched survivability of married women and unmarried women(together children). 

```python
married_women_prob_surv = []
unmarried_women_prob_surv = []


for idx in range(len(family_data)):
    married_data = family_data[idx].loc[family_data[idx]['Title'] == 'Mrs']['Survived'].values
    unmarried_data = family_data[idx].loc[family_data[idx]['Title'] == 'Miss']['Survived'].values
    
    for j in married_data:
        married_women_prob_surv.append(j)

    for k in unmarried_data:
        unmarried_women_prob_surv.append(k)
    
    
    #unmarried_women_prob_surv.append(family_data[idx].loc[family_data[idx]['Title'] == 'Miss']['Survived'])
    
    
    #single_woman = family_data[idx][family_data[idx]['Title'] == 'Miss']
    #unmarried_women.append(single_woman)

#survived_married_women = 0
#unsurvived_married_women = 0

#survived_unmarried_women = 0
#unsurvived_unmarried_women = 0

#married_women_prob_surv[0].empty()
```

```python
survived_married_women = married_women_prob_surv.count(1.00)
unsurvived_married_women = married_women_prob_surv.count(0.00)

survived_unmarried_women = unmarried_women_prob_surv.count(1.00)
unsurvived_unmarried_women = unmarried_women_prob_surv.count(0.00)

print("Survived married women: " + str(survived_married_women) + '\n' + 
      "Unsurvived married women: " + str(unsurvived_married_women) + '\n' + 
      "Survived unmarried women: " + str(survived_unmarried_women) + '\n' + 
      "Unsurvived unmarried women: " + str(unsurvived_unmarried_women))
```

Independently of families, it was analyzed all passengers and their titles. 

```python
import statsmodels.api as sm
```

Only Mr, Miss, and Mrs titles

```python
undesired_titles = ['Dr', 'Capt', 'Col', 'Don', 'Jonkheer', 'Lady', 'Major', 'Master', 'Mlle', 'Mme', 'Ms', 'Rev', 'Sir', 'the Countess']

for title in undesired_titles:
    new_data = new_data[new_data.Title != title]
```

```python
data = new_data[["Title", "Survived"]]
data.to_csv('/home/nbuser/library/titanic_r.csv', sep=',')

table = sm.stats.Table.from_data(data)
```

```python
print(table.table_orig)
```

So, this is surviving numbers of married and unmarried people(only for women). This table says that unmarried women and children survived on this disaster, mostly. Likewise, number of survived married women is more than unsurvived married women. Unfortunately, most of male passengers died on disaster. 

```python
table = sm.stats.Table.from_data(data)
print(table.test_nominal_association())
```

p-value says that this data cannot create randomly. Also, statistic is strong. 

```python
print(table.chi2_contribs)
```

```python
print(table.independence_probabilities)
```

Finally, this analysis says that there is a strong correlation between people who have 'Miss', 'Mr', 'Mrs' titles and surviving from this accident. Strongest correlation is between unmarried or married women and surviving.
